```python
import pandas as pd

df = pd.read_csv("data/clean_data.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

#Se aplica Label Encoding
label_encoder = LabelEncoder()
df['Accident_label'] = label_encoder.fit_transform(df['accident'])

#Mostrar valores asignados
print("\nAsignación de valores en Label Encoding:")
for clase, valor in zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)):
    print(f"{clase}: {valor}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#definimos las caracteristicas y la variable objetivo
x = df[['model_year', 'milage','accident']]
y = df[['price']]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)

print(f"Score en conjunto de prueba: {model.score(x_test, y_test)}")
```
